### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('loan_data.csv')
df
```

Removing ID Columns from DataSet and dropped Missing Values from Datset.

```python
df = df.drop(columns='Id')
df
```

```python
df.isnull().sum()
```

#### Removing all missing values

```python
df = df.dropna()
df
```

```python
df.info()
```

### Convert some Datatypes of the Columns 

```python
df['Family members'] = df['Family members'].astype(int)
df['Credit_History'] = df['Credit_History'].astype(int)

df.info()
```

```python
df.describe()
```

### Checking Outlier - in `Income of applicant`

```python
mx= df['Income of applicant'].max()
mn = df['Income of applicant'].min()

mx , mn
```

```python
q1,q3 = np.percentile(df['Income of applicant'], [25, 75])
q1,q3
```

```python
IQR = q3-q1
IQR
```

```python
lx = q1-1.5*IQR
ux = q3+1.5*IQR

lx , ux
```

```python
df = df.loc[(df['Income of applicant'] > lx) & (df['Income of applicant'] < ux)]

df
```

### Finding Oulier in column `Income of coapplicant`

```python
mx= df['Income of coapplicant'].max()
mn = df['Income of coapplicant'].min()

mx, mn
```

```python
q1, q3 = np.percentile(df['Income of coapplicant'], [25,75])

q1, q3
```

```python
IQR = q3 - q1 
IQR
```

```python
lx = q1-1.5*IQR
ux = q3 + 1.5*IQR

lx , ux 
```

```python
df = df.loc[(df['Income of coapplicant'] >lx) & (df['Income of coapplicant'] < ux)]

df
```

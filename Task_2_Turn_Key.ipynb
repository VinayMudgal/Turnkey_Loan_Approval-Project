### Encode categorical features 

```python
Label = LabelEncoder()
```

```python
df['Gender'] = Label.fit_transform(df['Gender'])
df['Marital Status'] = Label.fit_transform(df['Marital Status'])
df['Status of education'] = Label.fit_transform(df['Status of education'])
df['Family members'] = Label.fit_transform(df['Family members'])
df['Self_Employed'] = Label.fit_transform(df['Self_Employed'])
df['Property_Area'] = Label.fit_transform(df['Property_Area'])
```
